IA em saúde e relação médico-paciente

O notebook lê o dataset, padroniza variáveis, gera estatísticas descritivas e produz gráficos simples para o relatório.


```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

plt.rcParams['figure.figsize'] = (8, 4)
```

```python
df = pd.read_csv('artigos_ia_medico_paciente.csv')
print('Dimensões:', df.shape)
df.head()
```

```python
#alteração realizada 13/03/26: padronização dos nomes das colunas - Danilo Santos
df.columns = [c.strip().lower() for c in df.columns]
colunas_binarias = ['aborda_etica','aborda_confianca','aborda_empatia','aborda_comunicacao','aborda_autonomia','incluido_revisao']
for col in colunas_binarias:
    df[col] = df[col].astype(str).str.strip().str.lower().replace({'não':'nao'})
df['impacto_geral'] = df['impacto_geral'].astype(str).str.strip().str.lower()
df['tipo_ia'] = df['tipo_ia'].astype(str).str.strip().str.lower()
df['foco_principal'] = df['foco_principal'].astype(str).str.strip().str.lower()
df['ano'] = pd.to_numeric(df['ano'], errors='coerce')
df_incluidos = df[df['incluido_revisao'] == 'sim'].copy()
print('Artigos incluídos:', len(df_incluidos))
```

```python
print('\nDistribuição por ano:')
print(df_incluidos['ano'].value_counts().sort_index())
print('\nDistribuição por tipo de IA:')
print(df_incluidos['tipo_ia'].value_counts())
print('\nDistribuição por foco principal:')
print(df_incluidos['foco_principal'].value_counts())
print('\nDistribuição por impacto geral:')
print(df_incluidos['impacto_geral'].value_counts())
```

```python
#alteração realizada 15/03/26: filtragem dos artigos incluídos - Henrique Gomes
df_incluidos['ano'].value_counts().sort_index().plot(kind='bar')
plt.title('Quantidade de artigos por ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

```python
df_incluidos['tipo_ia'].value_counts().plot(kind='bar')
plt.title('Distribuição dos tipos de IA')
plt.xlabel('Tipo de IA')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

```python
df_incluidos['impacto_geral'].value_counts().plot(kind='bar')
plt.title('Impacto geral atribuído à IA')
plt.xlabel('Classificação')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

```python
temas = {
    'Ética': (df_incluidos['aborda_etica'] == 'sim').sum(),
    'Confiança': (df_incluidos['aborda_confianca'] == 'sim').sum(),
    'Empatia': (df_incluidos['aborda_empatia'] == 'sim').sum(),
    'Comunicação': (df_incluidos['aborda_comunicacao'] == 'sim').sum(),
    'Autonomia': (df_incluidos['aborda_autonomia'] == 'sim').sum(),
}
temas_series = pd.Series(temas).sort_values(ascending=False)
print(temas_series)
temas_series.plot(kind='bar')
plt.title('Temas abordados nos artigos')
plt.xlabel('Tema')
plt.ylabel('Quantidade de artigos')
plt.tight_layout()
plt.show()
```

```python
#alteração realizada 15/03/26: limpeza de texto - Amanda Massari
texto = ' '.join(df_incluidos['observacoes'].dropna().astype(str).tolist()).lower()
texto = re.sub(r'[^a-zA-ZÀ-ÿ\\s]', ' ', texto)
palavras = [p for p in texto.split() if len(p) > 3]
stopwords = {'sobre','para','com','uma','mais','como','entre','esse','essa','isso','pela','pelo','dos','das','nos','nas','que','por','sao','discute','avalia','analisa','foca'}
palavras_filtradas = [p for p in palavras if p not in stopwords]
frequencia = Counter(palavras_filtradas)
print(frequencia.most_common(10))
```
